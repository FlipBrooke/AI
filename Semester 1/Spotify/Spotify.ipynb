```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('spotify2024.csv', encoding='ISO-8859-1')
df["YouTube Views"] = df["YouTube Views"].str.replace(',', '')
df["YouTube Views"] = pd.to_numeric(df["YouTube Views"], errors='coerce')
type(df.loc[0]["YouTube Views"])
```

```python
list(df.columns)
```

```python
janSongs = 0
febSongs = 0
marSongs = 0
aprilSongs = 0
maySongs = 0
juneSongs = 0
julySongs = 0
augSongs = 0
sepSongs = 0
octSongs = 0
novSongs = 0
decSongs = 0


for song in range(len(df)):
    release_date = df.loc[song, 'Release Date']
    year = release_date[-4:]
    month = release_date.split("/")[0]
    
    if year == '2023':
        if month == '1':
            janSongs += 1
        elif month == '2':
            febSongs += 1
        elif month == '3':
            marSongs += 1
        elif month == '4':
            aprilSongs += 1
        elif month == '5':
            maySongs += 1
        elif month == '6':
            juneSongs += 1
        elif month == '7':
            julySongs += 1
        elif month == '8':
            augSongs += 1
        elif month == '9':
            sepSongs += 1
        elif month == '10':
            octSongs += 1
        elif month == '11':
            novSongs += 1
        elif month == '12':
            decSongs += 1
```

```python
dates = {
    "Months": ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    "Count": [janSongs, febSongs, marSongs, aprilSongs, maySongs, juneSongs, julySongs, augSongs, sepSongs, octSongs, novSongs, decSongs]
}

dates_df = pd.DataFrame(dates)

sns.barplot(x='Months', y='Count', data=dates_df)
```

```python
import pandas as pd
import seaborn as sns
```

```python
df_2023 = pd.read_csv('spotify2023.csv', encoding='ISO-8859-1')
df = pd.read_csv('spotify2024.csv', encoding='ISO-8859-1')
df["YouTube Views"] = df["YouTube Views"].str.replace(',', '')
df["YouTube Views"] = pd.to_numeric(df["YouTube Views"], errors='coerce')
type(df.loc[0]["YouTube Views"])

df["Spotify Streams"] = df["Spotify Streams"].str.replace(',', '')
df["Spotify Streams"] = pd.to_numeric(df["Spotify Streams"], errors='coerce')
type(df.loc[0]["Spotify Streams"])

df["Shazam Counts"] = df["Shazam Counts"].str.replace(',', '')
df["Shazam Counts"] = pd.to_numeric(df["Shazam Counts"], errors='coerce')
type(df.loc[0]["Shazam Counts"])


df["Soundcloud Streams"] = df["Soundcloud Streams"].str.replace(',', '')
df["Soundcloud Streams"] = pd.to_numeric(df["Soundcloud Streams"], errors='coerce')
type(df.loc[0]["Soundcloud Streams"])

df["YouTube Likes"] = df["YouTube Likes"].str.replace(',', '')
df["YouTube Likes"] = pd.to_numeric(df["YouTube Likes"], errors='coerce')
type(df.loc[0]["YouTube Likes"])

df["TikTok Likes"] = df["TikTok Likes"].str.replace(',', '')
df["TikTok Likes"] = pd.to_numeric(df["TikTok Likes"], errors='coerce')
type(df.loc[0]["TikTok Likes"])

df_2023["streams"] = df_2023["streams"].str.replace(',', '')
df_2023["streams"] = pd.to_numeric(df_2023["streams"], errors='coerce')
type(df_2023.loc[0]["streams"])

type(df.loc[0, "TikTok Likes"])
```

```python
janSongs = 0
febSongs = 0
marSongs = 0
aprilSongs = 0
maySongs = 0
juneSongs = 0
julySongs = 0
augSongs = 0
sepSongs = 0
octSongs = 0
novSongs = 0
decSongs = 0


for song in range(len(df)):
    release_date = df.loc[song, 'Release Date']
    year = release_date[-4:]
    month = release_date.split("/")[0]
    
    if year == '2023':
        if month == '1':
            janSongs += 1
        elif month == '2':
            febSongs += 1
        elif month == '3':
            marSongs += 1
        elif month == '4':
            aprilSongs += 1
        elif month == '5':
            maySongs += 1
        elif month == '6':
            juneSongs += 1
        elif month == '7':
            julySongs += 1
        elif month == '8':
            augSongs += 1
        elif month == '9':
            sepSongs += 1
        elif month == '10':
            octSongs += 1
        elif month == '11':
            novSongs += 1
        elif month == '12':
            decSongs += 1
```

```python
dates = {
    "Months": ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    "Count": [janSongs, febSongs, marSongs, aprilSongs, maySongs, juneSongs, julySongs, augSongs, sepSongs, octSongs, novSongs, decSongs]
}

dates_df = pd.DataFrame(dates)

sns.barplot(x='Months', y='Count', data=dates_df)
```

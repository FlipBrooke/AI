```python
from bs4 import BeautifulSoup as bs4
import requests
import re
```

```python
url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)
soup = bs4(response.text, 'html.parser')
h3_tags = soup.find_all('h3')
for h3 in h3_tags:
    print(h3.get_text(strip=True))
```

```python
aCount = 0
for h3 in h3_tags:
    text = h3.get_text(strip=True)
    if "a" in text.lower():
        aCount += 1
print(aCount)
print(len(text))
```

```python
import requests
from bs4 import BeautifulSoup as bs4

url = "https://en.wikipedia.org/wiki/Andorra"
response = requests.get(url)
soup = bs4(response.text, 'html.parser')

# Find all anchor tags with 'href' attribute
a_tags = soup.find_all('a', href=True)

# Print the href attribute of each anchor tag
for a in a_tags:
    print(a['href'])
```

```python
text = '''Your text containing Wikipedia links like: https://en.wikipedia.org/wiki/Python_(programming_language)'''

# Check if text is a string
if isinstance(text, str):
    # Define the pattern to match Wikipedia URLs
    pattern = r'https://en\.wikipedia\.org/wiki/[A-Za-z0-9_]+'

    wikipedia_links = re.findall(pattern, text)
    
    print(wikipedia_links)
else:
    print("Provided text is not a valid string")
```
